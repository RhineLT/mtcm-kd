```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Define the modalities and performance dictionary
modalities = ["T1ce", "T1", "T2", "FLAIR"]
performance_dict = { 
    "nnUNet": [81.7, 52.0, 65.3, 64.7],
    "UAGAN": [77.2, 55.7, 65.9, 66.7],
    "ReMIC": [75.9, 53.5, 59.5, 55.1],
    "Lie et al.": [82.4, 59.1, 69.6, 69.2],
    "Our Proposed": [90.4, 84.6, 61.1, 81.0]
}

# Create a dataframe from the performance dictionary
df = pd.DataFrame(performance_dict, index=modalities)

# Set the style of the plot
sns.set(style="whitegrid")

# Create the plot
plt.figure(figsize=(10, 6), dpi=300)  # Set the DPI to 300
sns.lineplot(data=df, markers=True, palette=["#f781bf", "#377eb8", "#ff7f00", "#a65628", "#e41a1c"])

# Set the labels and title
plt.xlabel("MRI Modalities")
plt.ylabel("Mean Dice Scores (%) of WT, TC, and ET Tumors")
plt.title("Performance Comparisons of Previous Models on Different Modalities using BraTS-2020 Dataset")

## change the legend position to top right
plt.legend(loc='upper right')

# Show the plot
plt.show()
```
